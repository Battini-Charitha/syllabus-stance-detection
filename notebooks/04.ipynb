# Notebook 04 — Annotation Validation

## Overview
This notebook documents the annotation validation process for the 354 manually labeled syllabus sentences across three stance categories: **discouraging**, **conditional**, and **encouraging**.

## Annotation Process
- All 354 sentences were manually annotated by a single annotator (Charitha Battini)
- Each sentence was reviewed to verify correct stance label assignment
- Edge cases particularly sentences on the boundary between **conditional** and **discouraging** were reviewed a second time to ensure consistency

## Label Definitions Used
| Label | Definition |
|-------|-----------|
| Discouraging (0) | Explicitly prohibits or discourages Wikipedia use |
| Conditional (1) | Permits use under specific conditions or qualifications |
| Encouraging (2) | Explicitly endorses or welcomes Wikipedia use |

## Validation Notes
- Label encoding is fixed: discouraging=0, conditional=1, encouraging=2
- Synthetic augmentation data (Llama-3 generated) was reviewed separately to ensure generated sentences matched their target label
- Test set (71 sentences) contains only real annotated sentences ever used during training

## Limitation
Annotation was performed by a single annotator. While internal consistency was verified through self-review, the absence of a second independent annotator means formal inter-annotator agreement (e.g. Cohen's Kappa) could not be computed. This is acknowledged as a limitation in the paper.
